<p style="background-color:#D1EEEE;color:black;font-size:23px;text-align:center;border-radius:10px 10px;border-style:solid;border-color:#1B1B1B;"> Overview of Chisquare Test of Independence  </p>


A Chi-Square Test of Independence is used to determine whether or not there is a significant association between two categorical variables. It is a nonparametric test
This test makes four assumptions:

* Assumption 1: Both variables are categorical.
* Assumption 2: All observations are independent.
* Assumption 3: Cells in the contingency table are mutually exclusive.(It’s assumed that individuals can only belong to one cell in the contingency table. That is, cells in the table are mutually exclusive – an individual cannot belong to more than one cell.)
* Assumption 4: Expected value of cells should be 5 or greater in at least 80% of cells.

<p style="background-color:#D1EEEE;color:black;font-size:23px;text-align:center;border-radius:10px 10px;border-style:solid;border-color:#1B1B1B;"> Age Range Vs Passenger's Satisfaction </p>


> * Null Hypothesis (H0) — Age range and satisfaction of passengers are independent of each other
> * Alternative Hypothesis (HA) — There is an association between the age range and satisfaction of passengers

```r
satisfied = df %>% 
filter(Satisfaction == 'Satisfied') %>% 
group_by(Age_Range) %>% 
summarize(satisfied = n())

dissatisfied = df %>% 
filter(Satisfaction == 'Neutral or Dissatisfied') %>% 
group_by(Age_Range) %>% 
summarize(dissatisfied = n())

joined_df1 = data.frame(satisfied, dissatisfied)
rownames(joined_df1) = c('Children','Middle Aged','Old','Youth')
joined_df1 = joined_df1 %>% select(satisfied, dissatisfied)
joined_df1

```

```r
# Chi-Square test of Independence

chisq.test(joined_df1) # n=129487
```

Since the p-value is less than the significance level(α = .05), we reject the null hypothesis at the 5% significance level.

> **There is sufficient evidence to support the claim that there is an association between age range and satisfaction level of passengers**

**Cramer's V**

Cramer’s V is a measure of the strength of association between two nominal variables.It’s appropriate to calculate V when you’re working with any table larger than a 2 x 2 contingency table.

The total sample size is n=129487, and we have already computed the Chi-Square statistics, 
X-squared = 4928.4
Let the number of rows be r = 4 and let c = 2 be the number of columns.

The Cramer's V statistic is computed as follows

Cramer’s V = √(X2/n) / min(c-1, r-1)

min(c-1,r-1)=min(2-1,4-1)=min(1,3)=1

```r
# Calculation of cramer's V

sqrt(4948.4 / 129487)
```

The following table shows how to interpret Cramer’s V based on the degrees of freedom:



The degrees of freedom would be calculated as:

df = min(#rows-1, #columns-1)
df = min(1, 3)
df = 1

Referring to the table above, we can see that a Cramer’s V of 0.19 and degrees of freedom = 1 indicates a **small (or “weak”) association between age range and satisfaction of passengers**.

**Phi (φ)**

It’s appropriate to calculate φ only when you’re working with a 2 x 2 contingency table (i.e. a table with exactly two rows and two columns).

Phi is calculated as φ = √(X-squared / n)

Here n= 129487 and X-squared = 4475.2

```r
# Calculation of phi

sqrt(4475.2 / 129487)
```

A value of φ  = 0.1 is considered to be a small effect, 0.3 a medium effect, and 0.5 a large effect.
This indicates a **small (or “weak”) association between type and satisfaction of passengers**.
